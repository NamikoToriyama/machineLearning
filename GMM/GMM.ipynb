```python
from sklearn.datasets import load_iris
from sklearn.mixture import GaussianMixture
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
data = load_iris()
n_components = 3 # クラスタ数を3に指定
model = GaussianMixture(n_components=n_components)
model.fit(data.data)
```

```python
print(model.predict(data.data)) # クラスを予測
print(model.means_) # 各ガウス分布の平均
print(model.covariances_) # 各クラス分布の分散
```

```python
n_components = 2 # 次元削減後の次元を２にする
model2 = PCA(n_components=n_components)
model2 = model2.fit(data.data)
transform_data = model2.transform(data.data) # 変換したデータ
data2 = list(zip(*transform_data))

plt.scatter(data2[0], data2[1], c=model.predict(data.data))
```
