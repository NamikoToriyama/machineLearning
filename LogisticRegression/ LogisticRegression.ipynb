# ロジスティック回帰

```python
import numpy as np
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt
import math
```

# シグモイド関数を描画する

```python
x = np.arange(-5, 5, 0.1)
e = math.e
# シグモイド関数
y = 1 / (1 + e**-x)

plt.plot(x, y)
plt.show()
```

```python
X_train = np.r_[np.random.normal(3, 1, size = 50), 
                            np.random.normal(-1, 1, size=50)].reshape((100, -1))
y_train = np.r_[np.ones(50), np.zeros(50)]
plt.scatter(X_train[:50], y_train[:50], c = 'orange')
plt.scatter(X_train[50:], y_train[50:], c = 'blue')

plt.grid()
plt.show() 
model = LogisticRegression()
model.fit(X_train, y_train)
model.predict_proba([[0], [1], [2]])[:, 1]
```

## 1℃の時に50%になるため、1℃を中心とした回帰直線をひく

```python
plt.scatter(X_train[:50], y_train[:50], c = 'orange')
plt.scatter(X_train[50:], y_train[50:], c = 'blue')
y = 1 / (1 + e**(-x+1)) 
plt.plot(x, y)
plt.grid()
plt.show()
```
