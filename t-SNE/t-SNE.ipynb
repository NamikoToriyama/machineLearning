```python
from sklearn.datasets import load_digits
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
from sklearn.manifold import LocallyLinearEmbedding
from sklearn.decomposition import PCA
```

```python
data = load_digits()
n_components = 2 # 削減後の次元数
model = TSNE(n_components=n_components)

transform_data = model.fit_transform(data.data)
print(transform_data)
```

```python
data2 = list(zip(*transform_data))

plt.scatter(data2[0], data2[1], c=data.target)
```

# PCAで描画

```python
n_components = 2 # 次元削減後の次元を２にする
model2 = PCA(n_components=n_components)
model2 = model2.fit(data.data)
transform_data = model2.transform(data.data) # 変換したデータ
data2 = list(zip(*transform_data))

plt.scatter(data2[0], data2[1], c=data.target)
```

# LLEで描画

```python
n_neighbors = 10 # 近傍点の数
n_components = 2 # 削減後の次元数
model = LocallyLinearEmbedding(n_neighbors=n_neighbors, n_components=n_components)

model.fit(data.data)
transform_data = model.transform(data.data)
data2 = list(zip(*transform_data))

plt.scatter(data2[0], data2[1], c=data.target)
```
