```python
# numpyを使って線形回帰を行うプログラム
import numpy as np
import matplotlib.pyplot as plt
```

# 一次で線形回帰を行う

```python
x = [10.0, 8.0, 13.0, 9.0, 11.0, 14.0, 6.0, 4.0, 12.0, 7.0, 5.0]
y = [8.04, 6.95, 7.58, 8.81, 8.33, 9.96, 7.24, 4.26, 10.84, 4.82, 5.68]
```

```python
print(np.polyfit(x, y, 1)) # 1次式で近似している
```

```python
plt.scatter(x, y)
plt.plot(x, np.poly1d(np.polyfit(x, y, 1))(x), color=(1,0,0))
plt.grid()
plt.show() 
```

# 最小２乗法を使った回帰(4次)

```python
# 参考: https://qiita.com/K_Yagi/items/3381e2357b4d50dc7f56
x_observed = [9, 28, 38, 58, 88, 98, 108, 118, 128, 138, 148, 158, 168, 178, 188, 198, 208, 218, 228, 238, 278, 288, 298]
y_observed = [51, 80, 112, 294, 286, 110, 59, 70, 56, 70, 104, 59, 59, 72, 87, 99, 64, 60, 74, 151, 157, 57, 83]


x_latent = np.linspace(min(x_observed), max(x_observed), 100)
fitted_curve = np.poly1d(np.polyfit(x_observed, y_observed, 4))(x_latent)
plt.scatter(x_observed, y_observed)
plt.plot(x_latent, fitted_curve, c="red", label="fitted")
plt.grid()
plt.show() 
```
